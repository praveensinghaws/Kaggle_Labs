# 1. Histogram
A histogram is a graphical representation of the distribution of numerical data. Here's how it works:

1. **Bins**: The range of data values is divided into intervals, or "bins," along the x-axis. Each bin represents a range of values.

2. **Frequency**: The frequency of data points falling within each bin is represented by the height of the corresponding bar on the y-axis. Essentially, it shows how many data points fall into each interval.

3. **Bars**: The bars in a histogram touch each other because the data is continuous and the intervals are adjacent. This helps visualize the overall shape of the distribution.

4. **Shape**: The shape of the histogram can provide insights into the underlying distribution of the data. Common shapes include symmetric (bell-shaped), skewed (positively or negatively), and uniform.

5. **Axis**: The x-axis represents the range of values or intervals, while the y-axis represents the frequency or count of data points in each interval.

Histograms are particularly useful for identifying patterns, outliers, and the overall spread of the data. They are commonly used in exploratory data analysis to gain insights into the characteristics of a dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
data = np.random.randn(1000)

# Create the histogram
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# Types of Skewed histograms :

Certainly, histograms can be categorized based on the symmetry of their distribution:

1. **Symmetric (Normal) Histogram**: This type of histogram displays a symmetric distribution, also known as a normal distribution. In a normal histogram, the data is evenly distributed around the mean, resulting in a bell-shaped curve. The left and right tails of the distribution are balanced, and the mean, median, and mode coincide at the center of the distribution.

2. **Right-Skewed (Positively Skewed) Histogram**: A right-skewed histogram has a longer right tail compared to the left tail. The majority of data points are concentrated on the left side of the histogram, with a few extremely high values extending the right tail. In this case, the mean is typically greater than the median.

3. **Left-Skewed (Negatively Skewed) Histogram**: Conversely, a left-skewed histogram exhibits a longer left tail and fewer data points on the right side. Most of the observations cluster towards the right side of the histogram, while a few exceptionally low values extend the left tail. Here, the mean is usually less than the median.

Each type of histogram distribution indicates different characteristics of the dataset, such as central tendency, dispersion, and skewness, providing valuable insights into the underlying data distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data for each distribution
data_normal = np.random.normal(loc=0, scale=1, size=1000)
data_right_skewed = np.random.exponential(scale=1, size=1000)
data_left_skewed = -np.random.exponential(scale=1, size=1000)

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot histograms
axs[0].hist(data_normal, bins=30, edgecolor='black', color='skyblue')
axs[0].set_title('Normal Distribution')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')

axs[1].hist(data_right_skewed, bins=30, edgecolor='black', color='salmon')
axs[1].set_title('Right Skewed Distribution')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Frequency')

axs[2].hist(data_left_skewed, bins=30, edgecolor='black', color='lightgreen')
axs[2].set_title('Left Skewed Distribution')
axs[2].set_xlabel('Value')
axs[2].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Create data for each distribution
data_normal = np.random.normal(loc=0, scale=1, size=1000)
data_right_skewed = np.random.exponential(scale=1, size=1000)
data_left_skewed = -np.random.exponential(scale=1, size=1000)

# Calculate mean, median, and mode for each distribution
mean_normal = np.mean(data_normal)
median_normal = np.median(data_normal)
try:
    mode_normal = stats.mode(data_normal).mode[0]
except IndexError:
    mode_normal = None

mean_right_skewed = np.mean(data_right_skewed)
median_right_skewed = np.median(data_right_skewed)
try:
    mode_right_skewed = stats.mode(data_right_skewed).mode[0]
except IndexError:
    mode_right_skewed = None

mean_left_skewed = np.mean(data_left_skewed)
median_left_skewed = np.median(data_left_skewed)
try:
    mode_left_skewed = stats.mode(data_left_skewed).mode[0]
except IndexError:
    mode_left_skewed = None

# Plot histograms
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot normal distribution
axs[0].hist(data_normal, bins=30, edgecolor='black', color='skyblue')
axs[0].axvline(mean_normal, color='red', linestyle='dashed', linewidth=2, label='Mean')
axs[0].axvline(median_normal, color='green', linestyle='dashed', linewidth=2, label='Median')
if mode_normal is not None:
    axs[0].axvline(mode_normal, color='orange', linestyle='dashed', linewidth=2, label='Mode')
axs[0].set_title('Normal Distribution')
axs[0].legend()

# Plot right skewed distribution
axs[1].hist(data_right_skewed, bins=30, edgecolor='black', color='salmon')
axs[1].axvline(mean_right_skewed, color='red', linestyle='dashed', linewidth=2, label='Mean')
axs[1].axvline(median_right_skewed, color='green', linestyle='dashed', linewidth=2, label='Median')
if mode_right_skewed is not None:
    axs[1].axvline(mode_right_skewed, color='orange', linestyle='dashed', linewidth=2, label='Mode')
axs[1].set_title('Right Skewed Distribution')
axs[1].legend()

# Plot left skewed distribution
axs[2].hist(data_left_skewed, bins=30, edgecolor='black', color='lightgreen')
axs[2].axvline(mean_left_skewed, color='red', linestyle='dashed', linewidth=2, label='Mean')
axs[2].axvline(median_left_skewed, color='green', linestyle='dashed', linewidth=2, label='Median')
if mode_left_skewed is not None:
    axs[2].axvline(mode_left_skewed, color='orange', linestyle='dashed', linewidth=2, label='Mode')
axs[2].set_title('Left Skewed Distribution')
axs[2].legend()

# Show plot
plt.tight_layout()
plt.show()
```

# Types of Histogram based on number of modes:

1. **Unimodal Histogram**: In a unimodal histogram, there is a single peak or mode, indicating that the data is predominantly concentrated around one central value. The distribution may be symmetric or skewed, but it has only one prominent peak.

2. **Bimodal Histogram**: A bimodal histogram displays two distinct peaks or modes, suggesting that the data has two separate groups or populations with different central tendencies. Each peak represents a distinct mode, indicating that the data is not uniformly distributed and has two primary clusters.

3. **Multimodal Histogram**: In contrast to unimodal and bimodal histograms, a multimodal histogram has three or more peaks or modes. This suggests that the data contains multiple subgroups or populations, each with its own central tendency. The presence of multiple modes indicates complex underlying patterns or structures within the dataset.

Identifying the number of modes in a histogram can provide insights into the underlying data distribution and help in understanding the distinct groups or patterns present in the dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data for unimodal histogram
unimodal_data = np.random.normal(loc=0, scale=1, size=1000)

# Generate sample data for bimodal histogram
bimodal_data = np.concatenate([np.random.normal(loc=-2, scale=1, size=500),
                                np.random.normal(loc=2, scale=1, size=500)])

# Generate sample data for multimodal histogram
multimodal_data = np.concatenate([np.random.normal(loc=-2, scale=1, size=300),
                                  np.random.normal(loc=0, scale=1, size=400),
                                  np.random.normal(loc=2, scale=1, size=300)])

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot unimodal histogram
axs[0].hist(unimodal_data, bins=30, color='skyblue', edgecolor='black')
axs[0].set_title('Unimodal Histogram')

# Plot bimodal histogram
axs[1].hist(bimodal_data, bins=30, color='salmon', edgecolor='black')
axs[1].set_title('Bimodal Histogram')

# Plot multimodal histogram
axs[2].hist(multimodal_data, bins=30, color='lightgreen', edgecolor='black')
axs[2].set_title('Multimodal Histogram')

# Show plot
plt.tight_layout()
plt.show()
```

# 2. Box Plot 

The box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a visual summary of key statistical measures, including the median, quartiles, and potential outliers. Here's how it works:

1. **Box**: The box in the plot represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3) of the dataset. The length of the box indicates the spread of the middle 50% of the data.

2. **Median**: The line inside the box represents the median (Q2) of the dataset, which is the middle value when the data is sorted in ascending order.

3. **Whiskers**: The whiskers extend from the edges of the box to the minimum and maximum values of the dataset that are within a certain distance from the quartiles. The distance is typically defined as 1.5 times the IQR.

4. **Outliers**: Any data points that fall beyond the whiskers are considered outliers and are plotted individually as points. These points represent values that are significantly different from the majority of the data and may warrant further investigation.

By visualizing the spread of the data and identifying potential outliers, box plots provide valuable insights into the distribution and variability of the dataset. They are particularly useful for comparing multiple groups or distributions side by side.

## Example 1 

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data with outliers
data_with_outliers = np.concatenate([np.random.normal(loc=0, scale=1, size=100), np.random.normal(loc=5, scale=1, size=10)])

# Create figure and axis objects
fig, axs = plt.subplots(1, 2, figsize=(10, 6))

# Box plot with outliers
axs[0].boxplot(data_with_outliers, vert=True, patch_artist=True, showfliers=True)
axs[0].set_title('With Outliers')

# Box plot without outliers
axs[1].boxplot(data_with_outliers, vert=True, patch_artist=True, showfliers=False)
axs[1].set_title('Without Outliers')

plt.tight_layout()
plt.show()
```

# Example 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data with outliers
data_with_outliers = np.concatenate([np.random.normal(loc=0, scale=1, size=100), np.random.normal(loc=5, scale=1, size=10)])

# Calculate quartiles
q1 = np.percentile(data_with_outliers, 25)
median = np.percentile(data_with_outliers, 50)
q3 = np.percentile(data_with_outliers, 75)

# Create figure and axis objects
fig, axs = plt.subplots(1, 2, figsize=(10, 6))

# Box plot with outliers
axs[0].boxplot(data_with_outliers, vert=True, patch_artist=True, showfliers=True)
axs[0].scatter(np.ones(len(data_with_outliers)), data_with_outliers, alpha=0.5)
axs[0].axhline(q1, color='g', linestyle='--', label='Q1')
axs[0].axhline(median, color='r', linestyle='--', label='Median')
axs[0].axhline(q3, color='b', linestyle='--', label='Q3')
axs[0].set_title('With Outliers')
axs[0].legend()

# Box plot without outliers
axs[1].boxplot(data_with_outliers, vert=True, patch_artist=True, showfliers=False)
axs[1].axhline(q1, color='g', linestyle='--', label='Q1')
axs[1].axhline(median, color='r', linestyle='--', label='Median')
axs[1].axhline(q3, color='b', linestyle='--', label='Q3')
axs[1].set_title('Without Outliers')
axs[1].legend()

plt.tight_layout()
plt.show()
```

# 3. Scatter Plots 

Scatter plots are indeed useful for visualizing the relationship between two continuous variables. Here's why they are valuable:

1. **Identifying Patterns**: Scatter plots allow us to quickly identify any patterns or relationships between two variables. By plotting each data point on a Cartesian plane, with one variable on the x-axis and the other on the y-axis, we can visually discern any trends or correlations.

2. **Assessing Correlation**: Scatter plots help us assess the strength and direction of the correlation between the two variables. If the points on the plot tend to form a clear linear pattern, it suggests a strong correlation. Conversely, if the points are scattered randomly, it indicates a weak or no correlation.

3. **Outlier Detection**: Scatter plots make it easy to identify outliers, which are data points that deviate significantly from the general trend. Outliers may indicate errors in data collection or provide insights into unusual observations that merit further investigation.

4. **Visualizing Clusters**: In addition to correlations, scatter plots can reveal clusters or groups within the data. These clusters may represent different categories or subgroups within the dataset, which can be further analyzed or explored.

Overall, scatter plots are a powerful tool in exploratory data analysis, allowing analysts to gain insights into the relationships and patterns present in their data.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
x = np.random.randn(100)
y1 = x + np.random.randn(100) * 0.5  # Identifying Patterns
y2 = 2 * x + np.random.randn(100) * 0.5  # Assessing Correlation
y3 = np.random.rand(100) * 10  # Outlier Detection
y4 = np.random.randn(100, 2)  # Visualizing Clusters

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Scatter plot 1: Identifying Patterns
axs[0, 0].scatter(x, y1)
axs[0, 0].set_title('Identifying Patterns')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')

# Scatter plot 2: Assessing Correlation
axs[0, 1].scatter(x, y2)
axs[0, 1].set_title('Assessing Correlation')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')

# Scatter plot 3: Outlier Detection
axs[1, 0].scatter(x, y3)
axs[1, 0].set_title('Outlier Detection')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

# Scatter plot 4: Visualizing Clusters
axs[1, 1].scatter(y4[:, 0], y4[:, 1])
axs[1, 1].set_title('Visualizing Clusters')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')

plt.tight_layout()
plt.show()
```

# Relationship

The relationship depicted in a scatter plot can be categorized into three main types:

1. **Positive Relationship (+ve)**: In a positive relationship, as one variable increases, the other variable also tends to increase. This is represented by a general upward trend in the scatter plot, indicating a positive correlation between the two variables. For example, as the amount of study time increases, exam scores also tend to increase.

2. **Negative Relationship (-ve)**: Conversely, in a negative relationship, as one variable increases, the other variable tends to decrease. This is depicted by a downward trend in the scatter plot, indicating a negative correlation between the two variables. An example of this could be the relationship between outdoor temperature and heating costs; as the temperature increases, heating costs tend to decrease.

3. **No Relationship**: Sometimes, there may be no apparent relationship between the two variables. In this case, the points on the scatter plot appear randomly scattered, with no discernible pattern or trend. This suggests that there is no correlation between the variables, and they are independent of each other. For instance, there might be no relationship between shoe size and intelligence quotient (IQ).

Understanding the nature of the relationship between variables is essential for making informed decisions and drawing meaningful conclusions from data analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Generate sample data
np.random.seed(0)
x = np.random.randn(100)
y_pos = 2 * x + np.random.randn(100)  # Positive Relationship
y_neg = -2 * x + np.random.randn(100)  # Negative Relationship
y_no = np.random.rand(100)  # No Relationship

# Introduce outliers for the fourth scatter plot
x_out = np.concatenate([x, [3, 4]])
y_out = np.concatenate([y_pos, [25, 20]])

# Create subplots
fig, axs = plt.subplots(1, 4, figsize=(20, 5))

# Scatter plot 1: Positive Relationship
axs[0].scatter(x, y_pos, color='blue')
axs[0].set_title('Positive Relationship')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Scatter plot 2: Negative Relationship
axs[1].scatter(x, y_neg, color='red')
axs[1].set_title('Negative Relationship')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')

# Scatter plot 3: No Relationship
axs[2].scatter(x, y_no, color='green')
axs[2].set_title('No Relationship')
axs[2].set_xlabel('X')
axs[2].set_ylabel('Y')

# Scatter plot 4: Outliers
axs[3].scatter(x_out, y_out, color='orange')
axs[3].set_title('Scatter Plot with Outliers')
axs[3].set_xlabel('X')
axs[3].set_ylabel('Y')

# Add regression lines to all scatter plots
for ax, x_data, y_data in zip(axs, [x, x, x, x_out], [y_pos, y_neg, y_no, y_out]):
    slope, intercept, _, _, _ = linregress(x_data, y_data)
    ax.plot(x_data, slope * x_data + intercept, color='black', linestyle='dashed')

plt.tight_layout()
plt.show()
```
